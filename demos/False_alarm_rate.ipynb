```python
import numpy as np
import multiprocessing
from multiprocessing import Pool
import utilizes_plot
"""
    Draws a random number from given probability density function.
    Parameters
    ----------
        pdf       -- the function pointer to a probability density function of form P = pdf(x)
        interval  -- the resulting random number is restricted to this interval
        pdfmax    -- the maximum of the probability density function
        integers  -- boolean, indicating if the result is desired as integer
        max_iterations -- maximum number of 'tries' to find a combination of random numbers (rand_x, rand_y) located below the function value calc_y = pdf(rand_x).
    returns a single random number according the pdf distribution.
"""
def draw_random_number_from_pdf(pdf, interval, pdfmax=1, integers=False, max_iterations=1000000):
    for i in range(max_iterations):
        if integers == True:
            rand_x = np.random.randint(interval[0], interval[1])
        else:
            # (b - a) * random_sample() + a
            rand_x = (interval[1] - interval[0]) * \
                np.random.random(1) + interval[0]
        rand_y = pdfmax * np.random.random(1)
        calc_y = pdf(rand_x)
        if(rand_y <= calc_y):
            return rand_x
    raise Exception("Could not find a matching random number within pdf in " +
                    max_iterations + " iterations.")
def pdf(x):
    mass_model=eval('utilizes_plot.hyper_prior_{}'.format(sub_model_name_list[11]) )
    return mass_model(x,*one_data_difmodel_maxl[str(mrms_data_label[2])+'_'+str(sub_model_name_list[11])] )

def draw_a_sample(x):
    return draw_random_number_from_pdf(pdf, [1.07, 2.4], pdfmax=3.9885, integers=False, max_iterations=1000000)[0]

def draw_samples(n_sample):
    xs = [0 for j in range(n_sample)]
    with Pool(64) as p:
        result = p.map(draw_a_sample, xs)
    return np.array(result)
n_sample = 200*87
phe_2G_false_data=draw_samples(n_sample)
plt.hist(phe_2G_false_data,bins=20,density=True,histtype='step')

mii=np.linspace(0.7,3,200)
mass_model_2g_fix=eval('utilizes_plot.hyper_prior_{}'.format(sub_model_name_list[11]) )
y_models_2g_fix=mass_model_2g_fix(mii,*one_data_difmodel_maxl[str(mrms_data_label[2])+'_'+str(sub_model_name_list[11])] ) 
plt.plot(mii,y_models_2g_fix ,label=r'{}{}'.format(model_name_label[11],data_name_labes[2]) )
plt.show()
```

```python
import numpy as np
from scipy.special import erf
from scipy.stats import beta as beta_dist
from scipy.stats import truncnorm
from scipy.interpolate import interp1d
import bilby
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
from bilby.core.prior import Uniform
from bilby.core.sampler import run_sampler
from bilby.core.prior import LogUniform
from bilby.hyper.likelihood import HyperparameterLikelihood
from scipy import interpolate
from scipy import integrate
import random
import argparse
import os
import glob
import warnings
import utilizes
warnings.filterwarnings("ignore")
#####################
outdir='false_alarm_outdir'
phe_2G_false_data_all=np.loadtxt('phe_2G_false_data_all.txt')
```

```python
lnevi=np.zeros(2)
for i in range(2):
    fal_dat=phe_2G_false_data_all[i*87:(i+1)*87]
    
    npool=10
    NN=int(len(fal_dat)/1)
    nlive=300
    mx=1
    #################################
    data_df=list()
    ns_m=abs(fal_dat)
    for j in range(NN):
        re=pd.DataFrame(ns_m[j*mx:(j+1)*mx],columns=['mu'])
        data_df.append(re)
    samples = data_df

    def run_prior(dataset):
        return 1/(2.9-0.9)

    hyper_prior=eval('utilizes.hyper_prior_{}'.format('turn_on_pow') )

    hp_likelihood = HyperparameterLikelihood(
        posteriors=samples, hyper_prior=hyper_prior,
        sampling_prior=run_prior, log_evidences=0, max_samples=mx)

    hp_priors =eval('utilizes.hp_priors_{}'.format('turn_on_pow') ) 

    # And run sampler
    result = run_sampler(
        likelihood=hp_likelihood, priors=hp_priors, sampler='dynesty', nlive=nlive,
        use_ratio=False, outdir=outdir, label='false_alarm_top_{}'.format(i),
        verbose=True, clean=True,npool=npool)
    lnevi[i] = result.log_evidence
np.savetxt('lnevi_top.txt',lnevi)
```
